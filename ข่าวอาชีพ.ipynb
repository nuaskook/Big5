```python
import bs4
import requests
import pandas as pd
import re
import openpyxl as xl
```

```python
topic_list = []
title_list = []
detail_list = []

data = requests.get(f'https://www.thansettakij.com/tags/%E0%B8%AD%E0%B8%B2%E0%B8%8A%E0%B8%B5%E0%B8%9E')
soup = bs4.BeautifulSoup(data.text)
    
for c in soup.find_all('div', class_='news-content'):
    topic_list.append(c.find('span').text.strip())
    title_list.append(c.find('h3').find('a').text.strip())
    detail_list.append(c.find('p', class_="--blurb").text.strip())
print('Complete')
    
table = pd.DataFrame([topic_list,title_list,detail_list]).transpose()
table.columns = ['topic','title','detail']
table.set_index('title')
```

```python
table.info()
```

```python
table.to_dict('ข่าวอาชีพ.csv')
```
