```python
from dsmlibrary.datanode import DataNode
import json
from tqdm.auto import tqdm
import pandas as pd
```

```python
dir_raw_id = 54
dir_process_id = 55
datanode = DataNode()
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='big.json'))
meta
```

```python
data = json.load(fp)
data
```

```python
spo = []
data = data
for news in tqdm(data):
        spo.append({
            'Picture': news.get('picture', ""),
            'Name': news.get('name', ""),
            'Twitter' : news.get('tHand'),
            'Conscious' : news.get('traits')['consc'],
            'Openness' : news.get('traits')['openn'],
            'Neuroticimn' : news.get('traits')['neuro'],
            'Agreeableness' : news.get('traits')['agree'],
            'Extroverion' : news.get('traits')['extra']
        })
df = pd.DataFrame(spo)
df
```

```python
datanode.write(df=df, directory=dir_process_id, name="big5", profiling=True, lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='big.json')])
```
